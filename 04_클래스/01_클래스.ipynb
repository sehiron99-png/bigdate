```python
result1 = 0

def add1(num):
    global result1 # 오류 찾기 어려움
    result1 += num
    return result1

print(add1(3))
print(add1(4))
print(add1(5))
```

```python
result2 = 0

def add2(num):
    global result2
    result2 += num
    return result2

print(add2(5))
print(add2(10))
print(add2(20))
```

```python
class Calculator: # class : 설계도 생성
    def __init__(self): # self <- 객체 본인 / init : 호출 안해도 자동 실행
        self.result = 0 # 초기화 값

    def add(self, num): # 추가
        self.result = self.result + num # ( 0 + 3 = 3 / 3 + 4 = 7)
        return self.result # return : 결과 값 밖으로
    
cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(3))
print(cal1.add(4))

print(cal2.add(5))
print(cal2.add(10))
```

```python
# 곱하기, 빼기, 나누기 기능 만들기
## mul() - 곱하기
## sub() - 빼기
## div() - 나누기

### 계산기 2개를 만들어서 (cal1, cal2)
### cal1 (10, 20)
### cal2 (20, 5)

class FourCal:
    def setdata(self, first, second):
        self.first = first    # 속성 값
        self.second = second  # 속성 값

    def add(self): # 함수
        result = self.first + self.second # self. : 변수 -> result : 저장
        return result

    def mul(self):
        result = self.first * self.second 
        return result
    
    def sub(self):
        result = self.first - self.second 
        return result
    
    def div(self):
        result = self.first / self.second 
        return result
    
cal1 = FourCal()
cal1.setdata(10, 20) # init 없어서 값 설정
print(cal1.add())
print(cal1.mul())
print(cal1.sub())
print(cal1.div())

cal2 = FourCal()
cal2.setdata(20, 5)
print(cal2.add())
print(cal2.mul())
print(cal2.sub())
print(cal2.div())
 

```

```python
class MoreFourCal(FourCal):
    def pow(self):
        result = self.first ** self.second
        return result
cal4 = MoreFourCal()
cal4.setdata(2, 4)
cal4.pow()         # 상속 받음 / 다른 파일 불러오기 import
```
