```python
# abs() - 절대값 만들어주는 함수
abs(-3) # 3

# all() - 요소가 모두 참이면 True
all([1, 2, 3]) # 0이 아니면 전부 True

# any() - 요소 중에 참이 하나라도 있다면 True
any([0, 1, 2, ""]) # True

# dir - 객체가 지닌 변수나 함수를 보여주는 함수
dir([1, 2, 3]) # 함수 목록 / 매서드 도움말 (빅데기 시험에 유용)

# id - 객체를 입력받아 객체의 고유 주소값을 반환하는 함수
a = 3
b = 4

print(id(a)) # 가르키고 있는 주소값
print(id(b))

# map(f, iterable)은 각 요소를 f함수에 적용한 결과를 보여주는 함수
def two_times(x):
    return x*2

list(map(two_times, [1, 2, 3, 4])) # [2, 4, 6, 8] 2배 값


```
