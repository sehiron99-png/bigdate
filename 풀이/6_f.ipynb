````python
# ### 문제 1: 인사 함수 만들기

# 이름을 입력받아 인사말을 출력하는 함수를 작성하세요.

# **요구사항**:

# - 함수 이름: `greet`
# - 매개변수: `name`
# - 반환값 없이 `print()`로 직접 출력

# **실행 예시**:

# ```
# greet("철수")
# → 안녕하세요, 철수님! 반갑습니다.

# greet("영희")
# → 안녕하세요, 영희님! 반갑습니다.
# ```

def greet(name):
    print(f"안녕하세요,{name}님! 반갑습니다.")
    
user = input("이름: ")
greet(user)
````

````python
# ### 문제 2: 사칙연산 함수

# 두 숫자와 연산자를 입력받아 계산 결과를 반환(return)하는 함수를 작성하세요.

# **요구사항**:

# - 함수 이름: `calculate`
# - 매개변수: `a`, `b`, `operator`
# - 지원 연산자: `+`, `-`, `*`, `/`
# - 0으로 나눌 경우 `"0으로 나눌 수 없습니다"` 반환
# - 잘못된 연산자 입력 시 `"잘못된 연산자입니다"` 반환

# **실행 예시**:

# ```python
# print(calculate(10, 3, "+"))   # 13
# print(calculate(10, 3, "-"))   # 7
# print(calculate(10, 3, "*"))   # 30
# print(calculate(10, 3, "/"))   # 3.3333333333333335
# print(calculate(10, 0, "/"))   # 0으로 나눌 수 없습니다
# print(calculate(10, 3, "%"))   # 잘못된 연산자입니다
# ```


def calculate(a,b,operator):
    if operator == "+":
        return a + b
    elif operator == "-":
        return a - b
    elif operator == "*":
        return a * b
    elif operator == "/":
        if b == 0:
            return "0으로 나눌 수 없습니다"
        return a / b
    else:
        return "잘못된 연산자입니다" 
num_1 = int(input("a: "))
num_2 = int(input("b: "))
op = input("operator(+, -, *, /): ")

result = calculate(num_1, num_2, op) # 1, 2, +
print(result)
````

````python
# ### 문제 3: 반환값이 여러 개인 함수

# 숫자 2개를 입력받아 합계와 평균을 동시에 반환하는 함수를 작성하세요.

# **요구사항**:

# - 함수 이름: `sum_and_avg`
# - 매개변수: `a`, `b`
# - 반환값: `(합계, 평균)` 튜플 형태

# **실행 예시**:

# ```python
# result = sum_and_avg(80, 90)
# print(result)          # (170, 85.0)

# s, a = sum_and_avg(75, 95)
# print(f"합계: {s}")    # 합계: 170
# print(f"평균: {a}")    # 평균: 85.0
# ```

def sum_and_avg(a, b):
    return a + b, (a + b)/2

num_1 = int(input("a: "))
num_2 = int(input("b: "))
result = sum_and_avg(num_1, num_2)
print(result)

s, a = sum_and_avg(num_1, num_2)
print(f"합계: {s}")
print(f"평균: {a}")
````

````python
# ### 문제 5: `*args` 활용 — 과목 점수 평균 (5분) ★★

# 여러 과목의 점수를 **개수 제한 없이** 입력받아 평균을 반환하는 함수를 작성하세요.

# **요구사항**:

# - 함수 이름: `get_average`
# - `*args`를 사용하여 가변 인수 처리
# - 입력값이 없으면 `0` 반환
# - 평균은 소수점 둘째 자리까지 반올림

# **실행 예시**:

# ```python
# print(get_average(80, 90, 100))        # 90.0
# print(get_average(75, 82, 91, 68, 95)) # 82.2
# print(get_average(100))                # 100.0
# print(get_average())                   # 0
# ```


def get_average(*args): 
    if len(args) == 0: 
        return 0
    average = sum(args) / len(args) # 총합 / 계수
    return round(average, 2)

score = []

while True:
    arg = input("점수: ")
    if arg == "x":
        break
    score.append(int(arg))
result = get_average(*score)
print(result)
````

````python
# ### 문제 6: 구구단을 출력하는 함수를 작성하세요.

# [요구사항]

# - 함수 이름: print_gugudan
# - 매개변수: dan (단), start=1 (시작 숫자), end=9 (끝 숫자)
# - for 문으로 구구단 출력
# - 잘못된 범위(start > end) 입력 시 "잘못된 범위입니다" 출력
# - 반환값: 없음 (print로 직접 출력)

# [실행 예시]

# print_gugudan(5)
# → 5 x 1 = 5
# 5 x 2 = 10
# ...
# 5 x 9 = 45

# print_gugudan(3, 3, 7)
# → 3 x 3 = 9
# 3 x 4 = 12
# 3 x 5 = 15
# 3 x 6 = 18
# 3 x 7 = 21

# print_gugudan(7, 9, 2)
# → 잘못된 범위입니다

def print_gugudan(dan, start = 1, end = 9):
    if start > end:
        return("잘못된 범위입니다")
    else:
        for i in range(start, end + 1): # [3, 4, 5, 6, 7]
            print(f"{dan} * {i} = {dan * i}")
        
print_gugudan(5)
print_gugudan(3, 3, 7)
print_gugudan(7, 9, 2)
````

````python
# ### 문제 7: 왼쪽 직각삼각형 

# 높이를 입력받아 왼쪽 정렬 직각삼각형을 출력하는 함수를 작성하세요.

# [요구사항]
# - 함수 이름: left_triangle
# - 매개변수: height
# - for 문 사용

# [실행 예시]

# left_triangle(5)

# *
# **
# ***
# ****
# *****


def left_triangle(height):
    for i in range(1, height + 1):
        print("*" * i)
        
left_triangle(5)
````
