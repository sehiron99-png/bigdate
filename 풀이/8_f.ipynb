```python
# 1. Product 클래스

class Product:
    """상품 정보를 관리하는 클래스"""

    def __init__(self, product_id, name, price, weight, stock):
        try:
            self.product_id = str(product_id)
            self.name = str(name)
            self.price = int(price)
            self.weight = float(weight)
            self.stock = int(stock)
        except ValueError:
            print("잘못된 데이터 타입입니다.")

        def get_info(self):
            """상품 정보를 딕셔너리로 반환"""
            return {
                "product_id": self.product_id,
                "name": self.name,
                "price": self.price,
                "weight": self.weight,
                "stock": self.stock
            }

        def update_stock(self, quantity):
            """
            재고 수량 업데이트
            재고가 부족하면 False 반환, 성공하면 True 반환
            """
            if self.stock >= quantity:
                self.stock -= quantity
                return True
            else:
                return False
            
        def apply_discount(self, discount_rate):
            """할인율을 적용한 가격 반환 (원본 가격은 유지)"""
            return int(self.price * (1 - discount_rate))

     
# ------------------------------------------------------------------------------

# 2. ShoppingCart 클래스 

class ShoppingCart:
    """장바구니를 관리하는 클래스"""
    
    def __init__(self):
        """장바구니 초기화"""
        self.items = []

    def add_product(self, product, quantity):
        """장바구니에 상품 추가"""
        if product.update_stock(quantity):
            self.items.append({ "product": product,"quantity": quantity})
            print(f"{product.name} {quantity}개 장바구니에 추가 완료")
        else:
            print(f"{product.name} 재고가 부족합니다.")
    
    def remove_product(self, product_id):
        """상품 ID로 장바구니에서 상품 제거"""
        for item in self.items:
            if item["product"].product_id == product_id:
                self.items.remove(item)
                print("상품이 장바구니에서 제거되었습니다.")
                return
        print("해당 상품이 장바구니에 없습니다.")

    def get_total_price(self):
        """장바구니 내 모든 상품의 총 금액 계산"""
        total_price = 0
        for item in self.items:
            total_price += item["product"].price * item["quantity"]
        return total_price
    
    def get_total_weight(self):
        """장바구니 내 모든 상품의 총 무게 계산"""
        total_weight = 0
        for item in self.items:
            total_weight += item["product"].weight * item["quantity"]
        return total_weight
    
    def show_cart(self):
        """장바구니 각 상품의 이름, 수량, 가격 표시"""
        for item in self.items:
            product = item["product"]
            quantity = item["quantity"]
            print(f"{Product.name} | 수량: {quantity} | 가격: {product.price * quantity}원")

        print(f"총 금액: {self.get_total_price()}원")
        print(f"총 무게: {self.get_total_weight()}kg")
        print("-" * 10)



# ------------------------------------------------------------------------------

# 3. Order 클래스 

class Order:
    """주문 정보를 관리하는 클래스"""
    
    def __init__(self, order_id, cart, customer_name):
        self.order_id = str(order_id)
        self.cart = cart
        self.customer_name = str(customer_name)
    
    def calculate_final_price(self):
        """최종 결제 금액 계산 (상품 금액 + 배송비)"""
        product_total = self.cart.get_total_price()
        shopping_fee = 0 if product_total >= 50000 else 3000
        return product_total + shopping_fee
    
    def print_receipt(self):
        """주문서 출력"""
        print("\n===== 주문서 =====")
        print(f"주문 ID: {self.order_id}")
        print(f"고객명: {self.customer_name}")
        
        print("\n주문 상품 목록: ")
        for item in self.cart.items:
            product = item["product"]
            quantity = item["quantity"]
            print(f"{product.name} ({quantity}개) - {product.price * quantity}원")

        product_total = self.cart.get_total_price()

        if product_total >= 50000:
            Shopping_fee = 0
        else:
            Shopping_fee = 3000
            
        final_price = product_total + Shopping_fee

        print("\n상품 금액: ", product_total, "원")
        print("배송비: ", Shopping_fee, "원")
        print("최종 결제 금액: ", final_price, "원")
        print("=" * 10)

        
```

```python
class Product:
    """상품 정보를 관리하는 클래스"""

    def __init__(self, product_id, name, price, weight, stock):
        # 생성자: 객체가 만들어질 때 실행됨
        self.product_id = product_id # item의 속성
        self.name = name
        self.price = price
        self.weight = weight
        self.stock = stock

    def get_info(self):
        """상품 정보를 딕셔너리로 반환"""
        return {
            "product_id": self.product_id,
            "name": self.name,
            "price": self.price,
            "weight": self.weight,
            "stock": self.stock
        }

    def update_stock(self, quantity):
        """
        재고 수량 업데이트
        재고가 부족하면 False, 성공하면 True
        """
        if self.stock >= quantity:
            self.stock -= quantity
            return True
        else:
            return False

    def apply_discount(self, discount_rate):
        """할인율을 적용한 가격 반환 (원본 가격은 유지)"""
        discounted_price = self.price * (1 - discount_rate)
        return int(discounted_price)
```

```python
# Order 클래스 구현
class Order:
    """주문 정보를 관리하는 클래스"""

    def __init__(self, order_id, cart, customer_name): # (self, 변수1, 변수2 ...)
        self.order_id = order_id # 변수
        self.cart = cart
        self.customer_name = customer_name

    def calculate_final_price(self):
        """
        최종 결제 금액 계산 (상품 금액 + 배송비)
        배송비는 총 금액이 100,000원 미만이면 3,000원
        """
        total_price = self.cart.get_total_price()
        #shipping_fee = 0 if total_price >= 100000 else 3000
        if total_price >= 100000:
            shipping_fee = 0
        else:
            shipping_fee = 3000
        return total_price + shipping_fee, shipping_fee

    def print_receipt(self): # self : order_id
        """주문서 출력"""
        print("\n===== 주문서 =====")
        print(f"주문 ID: {self.order_id}") # 변수
        print(f"고객명: {self.customer_name}")
        print("\n주문 상품 목록:")

        for item in self.cart.items: # self : order_id
            product = item["product"] # item : 딕셔너리
            quantity = item["quantity"]
            print(f"- {product.name} x {quantity}개")
            # {product(객체의).name(이름값)}

        total_price = self.cart.get_total_price() # self : order_id
        final_price, shipping_fee = self.calculate_final_price()
        # calcu..price는 두개를 받은 값

        print(f"\n상품 금액: {total_price}원")
        print(f"배송비: {shipping_fee}원")
        print(f"최종 결제 금액: {final_price}원")
```
