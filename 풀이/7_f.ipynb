````python
# 세 과목의 평균을 계산하는 함수
def calculate_average(korean, english, math):
    return(korean + english + math) / 3


# 성적 정보를 파일에 저장하는 함수
def grade_save(name, korean, english, math, average):
    data = f"{name},{korean},{english},{math},{round(average,2)}\n"
    with open("grades.txt", "a", encoding="utf-8") as f:
        f.write(data)
        # f.write(f"{name},{korean},{english},{math},{round(average,2)}\n")
        # -> 요약 가능

    

# 파일에서 성적을 읽어 출력하는 함수
def grade_file():
    print("\n===== 전체 성적 =====")
    with open("grades.txt", "a", encoding="utf-8") as f:
        for line in f:
            name, korean, english, math, average = line.strip().split(",")
            print(f"{name}: 국어 {korean}, 영어 {english}, 수학 {math} (평균: {average})")

    print("===================")
    
while True:
    name = input("학생 이름: ")
    korean = int(input("국어 점수: "))
    english = int(input("영어 점수: "))
    math = int(input("수학 점수: "))

    average = calculate_average(korean, english, math)
    print(f"평균: {average}점")

    grade_save(name, korean, english, math, average)
    print("\n성적이 저장되었습니다!")

    add = input("다른 학생 추가? (y/n): ")
    print()

    if add != "y":
        break

grade_file()
````
