````python
### 문제 7: 소수 판별기 
사용자가 입력한 숫자가 소수인지 판별하고, 1부터 해당 숫자까지의 모든 소수를 출력하는 프로그램을 작성하세요.

**요구사항**:
- for 문과 if 문 사용
- 2 이상의 숫자만 입력 가능
- 소수 판별 로직 구현

**실행 예시**:
```
숫자를 입력하세요: 20
20은(는) 소수가 아닙니다.

1부터 20까지의 소수:
2, 3, 5, 7, 11, 13, 17, 19
총 8개의 소수가 있습니다.
```
````

````python
n = int(input("숫자를 입력하세요: ")) # n : 사용자가 입력한 숫자 (20)
if n <2:                             # 에러 처리
    print("2 이상의 숫자만 입력 가능합니다.")
else:
    prime = True
    for num in range(2, n): # num : range(2, 20)-> [2,3,4,...19]
        if n % num == 0: # 검사할수 / 나눌수
            prime = False
            break

    if prime: # prime은 True/False (bool값)
        print(f"{n}은 소수입니다.")
    else:
        print(f"{n}은 소수가 아닙니다.")

# 1 ~ n까지 모든 소수 출력
    primes = []
    for i in range(2, n + 1): # range(2, 21) -> [2,3,4,...20]
        prime = True
        for d in range(2, i): # [2, 3] 실행
            if i % d == 0: # d = 2
                prime = False
        if prime == True: # if prime
            primes.append(i) # -> primes = [2, 3]

    print(f"1부터 {n}까지의 소수: ")

    print(primes)
    print(len(primes))
````

````python
### 문제 8: 간단한 ATM 시스템
간단한 ATM 기능을 구현하세요.

**요구사항**:
- while 문으로 메뉴 반복
- 초기 잔액: 10,000원
- 기능: 1.잔액조회 2.입금 3.출금 4.종료
- 출금 시 잔액 부족 체크

**실행 예시**:
```
=== ATM 시스템 ===
1. 잔액 조회
2. 입금
3. 출금
4. 종료
선택: 1
현재 잔액: 10,000원

선택: 2
입금액: 5000
입금 완료! 현재 잔액: 15,000원

선택: 3
출금액: 20000
잔액이 부족합니다. 현재 잔액: 15,000원

선택: 4
이용해 주셔서 감사합니다.
```
````

````python
money = 10000
print("""=== ATM 시스템 ===
1. 잔액조회
2. 입금
3. 출금
4. 종료""")

while True: # 맞을 시 계속 반복
    choice = input("선택: ")
    if choice == "1":
        print(f"현재 잔액: {money}원")

    elif choice == "2":
        plus = int(input("입금액: "))
        if plus <= 0:
            print("올바른 금액을 입력하세요")
        else:
            money += plus 
            print(f"입금 완료! 현재 잔액: {money}원")

    elif choice == "3":
        minus = int(input("출금액: "))
        if minus > money:
            print(f"잔액이 부족합나다. 현재 잔액: {money}원")
        elif minus <= 0:
            print("올바른 금액을 입력하세요")
        else:
            money -= minus
            print(f"출금 완료! 현재 잔액: {money}원")

    elif choice == "4":
        print("이용해 주셔서 감사합니다.")
        break

    else:
        print("잘못된 번호 입니다.")
````
