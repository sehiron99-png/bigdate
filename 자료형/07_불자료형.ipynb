```python
# 불 자료형은 어떻게 만들까?
a = True
b = False
print(a)
print(b)
print(type(a))
print(type(b))
```

```python
# 자료형의 참과 거짓 (0, 빈 값은 무조건 거짓)
print(bool("python"))
print(bool(""))

print(bool([1, 2, 3]))
print(bool([]))

print(bool((1, 2, 3)))
print(bool(()))

print(bool({"a" : 1}))
print(bool({}))

print(bool(1))
print(bool(0))

print(bool(None))
```

```python
# and 연산자, or 연산자
print(True and True)
print(True and False)
print(False and True)
print(False and False)

print(True or True)
print(True or False)
print(False or True)
print(False or False)

```

```python
# if 조건식 :
# 수행할 문장1
if 0: # 0은 False라 안됨
    print("0")
if 1:
    print("0")
```
