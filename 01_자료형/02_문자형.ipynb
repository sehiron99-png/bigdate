```python
# 역슬래시를 이용해서 작은 따옴표와 큰 따옴표를 문자열에 포함
food = "python\'s favorite \n is \t perl"
print(food)

# \n 줄 바꿈
# \t 탭 간격
# \\ \그대로 사용
# \' 작은 따옴표 그대로 사용
```

```python
# 문자열 인덱싱 (0부터 시작, 공백포함 x)
a = "Life is too short"
print(a[8])
print(a[-1])
print(a[-2])
```

```python
# 문자열 슬라이싱 (컨슆알트)
a = '''Life is too short, you need python'''
print(a[0:4])
print(a[12:17])
print(a[5:])
print(a[:17])
print(a[:])
```

```python
# 문자열 변수 출력 (약간 매크로 같은 느낌)
day = "three"
print(f"I was sick for {day} days")
```
