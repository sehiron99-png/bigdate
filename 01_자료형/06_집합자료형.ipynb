```python
# 집합 자료형은 어떻게 만들까?
s1 = set([1, 2, 3])
s2 = set("hello")
s3 = {1, 2, 3}
s4 = {}
print(s1)
print(s2)
print(s3)
print(s4)
print(type(s3))
print(type(s4))
```

```python
# 교집합, 합집합, 차집합 구하기
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])

## 교집합
print("교집합 : ", s1 & s2)

## 합집합 | (← 쉬프트\)
print("합집합 : ", s1 | s2 )

## 차집합
print("차집합 : ", s1 - s2 )
print("차집합 : ", s2 - s1 )
```

```python
# 집합 자료형 관련 함수
s1 = set([1, 2, 3])
print(s1)

## 값 1개 추가하기 - add
s1.add(4)
print(s1)

## 값 여러개 추가하기 - update
s1.update([4, 5, 6])
print(s1)

## 특정 값 제거하기 - remove
s1.remove(2)
print(s1)

## 모든 값 제거하기
s1.clear()
print(s1)

```
