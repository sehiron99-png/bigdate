```python
f = open("data2.txt", "w") # r: 읽기 w: 쓰기 f: 파일 객체 (아무거나 써도 됨)
f.close()

# 파일객체 = open(파일이름, 파일열기모드)
# r - 읽기 모드: 파일을 읽기만 할 때
# w - 쓰기 모드: 파일의 내용을 쓸 때
# a - 추가 모드: 파일 마지막에 내용을 추가할 때
```

```python
f = open("./data/새파일.txt", "w") # ./ : 현재 폴더 ★ (상대 경로)

for i in range(1, 10+1):
    data = f"{i}번째 줄입니다.\n"  # \n : 줄 바꿈
    f.write(data)                 # 반복문

f. close()                        # 닫는게 프로그램에게 좋음 (충돌 방지)
```

```python
f = open("data/새파일.txt", "r")
while True:    # 여러줄 읽기 = readlines
    line = f.readline()   # 라인은 이것
    if not line: break    # 더이상 없다면 닫기
    print(line)

f.close
```

```python
# readlines() 함수 이용하기
f = open("data/새파일.txt", "r")
lines = f.readlines()    # 여러줄 읽기

print(lines)
for line in lines:       # for : 반복문
    print(line)          # ,end="" 들여쓰기 1줄 x, \n 1줄 추가 / sep : 칸
f.close()

# print(type(lines))
# print(len(lines))
```

```python
# read 함수 사용하기
f = open("data/새파일.txt","r")
data = f.read()
print(data)
f.close()
```

```python
# 파일에 새로운 내용 추가하기
f = open("../새파일.txt", "a")    # ../ : 상위 폴더로 이동

for i in range(11, 20+1):
    f.write(f"{i}번째 줄입니다.\n")
f.close()
```

```python
# with 문과 함께 사용하기
with open("data3.txt","w") as f: # as : with 의 별칭
    f.write("1번째 줄입니다.\n")
```
