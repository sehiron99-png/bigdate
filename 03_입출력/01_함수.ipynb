```python
# 함수란 무엇인가?
## 믹서기, 자판기

# 파이썬 함수의 구조

# def 함수이름(매개변수):
#     수행할문장1
#     수행할문장2
#     ...

def add(a, b): # add가 받을 값
    return a + b # 더한 값을 밖으로 내보냄 (더하는 함수)
# 중간 계산 여러번 할 시 result a + b가 깔끔
# 결과 값은 return으로 가져옴

result = add(3, 4) # 결과 값 임시 보관
print(result)
```

```python
# 일반적인 함수
## 입력값: O, 반환값: O

# def 함수이름(매개변수):
#     수행할문장1
#     수행할문장2
#     ...
#     return 결과값

# 예제
def add(a, b):
    result = a + b # 계산 결과를 result에 임시 보관
    return result # return으로 add(3, 4)을 꺼내서 변수로 사용 가능

a = add(3, 4)
print(a)
```

```python
# 입력값이 없는 함수
## 입력값: X, 반환값: O
def say():
    return "Hi"

a = say()
print(a)
```

```python
# 반환값이 없는 함수
## 입력값: O, 반환값: X

def add(a, b):
    print(f"{a}, {b}의 합은 {a + b}이다")

a = add(3, 4) # 함수가 print만 있어서 나옴
print(a)
```

```python
# 입력값과 반환값이 없는 함수
## 입력값: X, 반환값: X

def say():
    print("Hi")

a = say() # () 안이 비어있음
print(a)  # 함수가 끝나서 none 출력
```

```python
# 입력값이 몇 개가 될지 모를 때는 어떻게 해야 할까?

# def 함수이름(*args):
#     수행할문장1
#     수행할문장2
#     ...

def add_many(*args): # 임외 입력값 모두 받음 *args = [1, 2, 3, 4, 5]
    result = 0       # 0은 초기화 느낌
    for i in args:
        result = result + i
    return result

a = add_many(1, 2, 3, 4, 5)
print(a)

print("Hi", "Hello",)
```

```python
# 키워드 매개변수, kwargs
## 키워드 값의 형태로 매개변수를 받을 때 (매개변수: 함수 호출시의 변수)
## 매개변수 앞에 **(2개)를 붙인다 (2개는 : 딕셔너리) / (1개는 , 튜플)

def print_kwargs(**kwargs):
    print(kwargs)

print_kwargs(a=1) # 딕셔너리 형태
print_kwargs(name="foo", age=3)
```

```python
def create_profile(**profiles):
    print("===== 프로필 정보 =====")
    for key, value in profiles.items(): # 리스트 형식
        print(f"{key}: {value}")

create_profile(이름="김철수", 나이=30, 직업="개발자", 취미="운동")
    
```

```python
# *args, **kwargs 혼용
# *args = 나머지 위치 인자를 튜플로
# **kwargs = 나머지 키워드 인자를 딕셔너리로
def mixed_function(name, *args, **kwargs):
    print(f"이름 : {name}")
    print(f"추가인수들 : {args}")
    print(f"추가키워드 : {kwargs}")

mixed_function("홍길동", 1, 2, 3, 4, 5, age=25, city="서울") # 호출
```

```python
def add_and_mul(a, b):
    return a+b
    return a*b # 리턴은 끝이라는 의미, 두가지면 줄바꿈 안함

result = add_and_mul(3, 4)
print(result)
```

```python
def say_nick(nick):
    if nick == "바보":
        return # 아무 값 없어도 실행됨 (종료, 에러 처리할 때 사용)
    print(f"나의 별명은 {nick}입니다.")
# say_nick("야호") # -> 야호
say_nick("바보") # -> 실행 안됨
```

```python
# 매개변수에 초깃값 미리 설정하기
def say_myself(name, age, man=True):
    print(f"나의 이름은 {name}입니다")
    print(f"나이는 {age}살 입니다")
    if man == True : # if man: 인수값에 의해 결정돼서 vsc에선 생략 가능
        print("남자 입니다.")
    else:
        print("여자 입니다.")

say_myself("홍길동", 30) # 정해져 있는 bool문은 생략할 수 있음
say_myself("영희", 30, False)
    
```

```python
# lambda 예약어
# def 함수이름 (매개변수):
#     수행할문장
#     return

# def add(a, b):
#     return a+ b

# lambda
# 함수이름 = lambda 매개변수1, 매개변수2, ... : 수행할문장

add = lambda a, b : a + b # 결과 반환 하는 return없어도 자동 반환
result = add(3, 4)        # add = 함수 이름
print(result)             # def : 여러줄, 이름o / lam : 한줄, 이름x
```

```python
def add(a, b):
    """
    함수의 독스트링(Docstring)
    두 숫자를 더하는 함수
    설명: 다른 사람 또는 나를 위한 문서
    """

    return a + b

result = add(3, 4)
print(result)

print(add.__doc__) # _ 2개
```
