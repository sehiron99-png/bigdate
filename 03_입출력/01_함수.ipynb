```python
# 함수란 무엇인가?
## 믹서기, 자판기

# 파이썬 함수의 구조

# def 함수이름(매개변수):
#     수행할문장1
#     수행할문장2
#     ...

def add(a, b): # add가 받을 값
    return a + b # 더한 값을 밖으로 내보냄 (더하는 함수)
# 중간 계산 여러번 할 시 result a + b가 깔끔
# 결과 값은 return으로 가져옴

result = add(3, 4) # 결과 값 임시 보관
print(result)
```

```python
# 일반적인 함수
## 입력값: O, 반환값: O

# def 함수이름(매개변수):
#     수행할문장1
#     수행할문장2
#     ...
#     return 결과값

# 예제
def add(a, b):
    result = a + b # 계산 결과를 result에 임시 보관
    return result # return으로 add(3, 4)을 꺼내서 변수로 사용 가능

a = add(3, 4)
print(a)
```

```python
# 입력값이 없는 함수
## 입력값: X, 반환값: O
def say():
    return "Hi"

a = say()
print(a)
```

```python
# 반환값이 없는 함수
## 입력값: O, 반환값: X

def add(a, b):
    print(f"{a}, {b}의 합은 {a + b}이다")

a = add(3, 4) # 함수가 print만 있어서 나옴
print(a)
```

```python
# 입력값과 반환값이 없는 함수
## 입력값: X, 반환값: X

def say():
    print("Hi")

a = say() # () 안이 비어있음
print(a)  # 함수가 끝나서 none 출력
```

```python
# 입력값이 몇 개가 될지 모를 때는 어떻게 해야 할까?

# def 함수이름(*args):
#     수행할문장1
#     수행할문장2
#     ...

def add_many(*args): # 임외 입력값 모두 받음 *args = [1, 2, 3, 4, 5]
    result = 0       # 0은 초기화 느낌
    for i in args:
        result = result + i
    return result

a = add_many(1, 2, 3, 4, 5)
print(a)

print("Hi", "Hello",)
```
